```python
def calculate_average_mean_fare(fares):
    if not fares:
        return 0.0
    
    sum_of_fares = sum(fares)
    number_of_fares = len(fares)
    
    average_mean_fare = sum_of_fares / number_of_fares
    return average_mean_fare


fares = [10, 15, 12, 17, 20]
average_mean_fare = calculate_average_mean_fare(fares)
print(f"The average mean fare is: ${average_mean_fare:.2f}")
```

```python
import matplotlib.pyplot as plt
from collections import Counter
import seaborn as sns

set1 = ['Standard', 'Standard', 'VIP', 'Economy', 'VIP']
set2 = ['VIP', 'Standard', 'Standard', 'Economy']
set3 = ['Standard', 'Economy', 'Economy', 'Economy']


count_set1 = Counter(set1)
count_set2 = Counter(set2)
count_set3 = Counter(set3)


ticket_types = list(count_set1.keys())  
counts_set1 = [count_set1[t] for t in ticket_types]
counts_set2 = [count_set2[t] for t in ticket_types]
counts_set3 = [count_set3[t] for t in ticket_types]



    

bar_width = 0.2
index = range(len(ticket_types))

plt.bar(index, counts_set1, width=bar_width, label='Set 1')
plt.bar([i + bar_width for i in index], counts_set2, width=bar_width, label='Set 2')
plt.bar([i + 2 * bar_width for i in index], counts_set3, width=bar_width, label='Set 3')

plt.xlabel('Ticket Types')
plt.ylabel('Counts')
plt.title('Counts of Different Ticket Types in Each Set')
plt.xticks([i + bar_width for i in index], ticket_types)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
  
home_destinations = {
    'City A': 1000000,
    'City B': 500000,
    'City C': 800000,
    'City D': 300000,
    'City E': 1200000
}


most_populated = max(home_destinations, key=home_destinations.get)
max_population = home_destinations[most_populated]


least_populated = min(home_destinations, key=home_destinations.get)
min_population = home_destinations[least_populated]


print(f"The most populated destination is {most_populated} with a population of {max_population}")
print(f"The least populated destination is {least_populated} with a population of {min_population}")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv(r"C:\Users\Satyug\Downloads\titanic.csv")
```

```python

embarked_mapping = {'C': 'Cherbourg', 'S': 'Southampton', 'Q': 'Queenstown'}
df['embarked_full'] = df['embarked'].map(embarked_mapping)


corr_survived_embarked = df.groupby('embarked_full')['survived'].mean()

print("Correlation of Survived with Embarked:")
print(corr_survived_embarked)
```

```python

corr_embarked_pclass = df.groupby('embarked_full')['pclass'].value_counts(normalize=True).unstack()

print("\nCorrelation of Embarked with Pclass:")
print(corr_embarked_pclass)
```
